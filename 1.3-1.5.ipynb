# 1.4 Pegasos

```python
features1 = features
```

```python
sentiment1 = sentiment
```

```python
LAMBDA, B = 0.001, 100
w = np.zeros((1, len(features1[0])))
T = 1000
errors = np.zeros(T, )
s = sentiment
```

```python
#regularization factor: lamda=0.001
#Number of iterations: T=1000
#Initialize w as zero vector
#Take a mini-batch with size B=100 from training set

for t in range(1, T): 
    #calculate the learning rate
    eta = 1 / (t * LAMBDA)
    
    #Take a mini-batch with size B
    selected_index = np.random.randint(low = 0, high = len(features1) - 1, size = B)
    
    index = np.random.randint(low = 0, high = len(features1) - 1, size = 100)
    predict = 2 * (features1[index].dot(w.T) >= 0) - 1
    errors[t - 1] = np.mean(s[index].reshape(100, 1) != predict)
   
    i = (s[selected_index].reshape(B, 1) * (features1[selected_index].dot(w.T))) < 1
    su = np.sum(i * s[selected_index].reshape(B, 1) * features1[selected_index], axis=0)
    
    #stochastic gradient
    delta = LAMBDA * w - eta / B * su
    w_prime = w - eta * delta
    
    #compute the projection
    try:
        w = min(1, 1 / (LAMBDA**0.5 * np.linalg.norm(w_prime, 2))) * w_prime
    except:
        w = w_prime
```

```python
import matplotlib.pyplot as plt
plt.plot(errors)
plt.title('Pegasos: Training Errors for 1000 iterations')
plt.xlabel('Number of Iterations')
plt.ylabel('Training Errors')
plt.show()
```

```python
np.save("training_errors_1000_pegasos.npy", errors)
```

# 1.5 AdaGrad

```python
features2 = features
sentiment2 = sentiment
```

```python
#Initial learning rate: eta=0.001
#Number of iterations: T=5000
#Initialize w as zero vector
#Initialize st as identity vector
#Take a mini-batch of size B=100 from training set

eta, p, B, lamda = 0.001, len(features2[0]), 100, 0.001
w, st = np.zeros((1, p)), np.ones(p, )
T = 5000
errors_ada = np.zeros(T, )
```

```python
for t in range(1, T):
    selected_index = np.random.randint(low = 0, high = len(features2) - 1, size = B)
    i = (sentiment2[selected_index].reshape(B, 1) * (features2[selected_index].dot(w.T))) < 1
    delta = np.sum(i * sentiment2[selected_index].reshape(B, 1) * features2[selected_index], axis = 0) / B
    
    #compute the square of gradient
    st = st + np.power(delta, 2)
    
    #compute square root of diagonal matrix G                   
    G = st  ** 0.5
    G_inverse = 1 / G
    
    #update the learning rate
    eta = 1 / (lamda * t)
    
    #compute Mahalanobis projection
    w_prime = w - eta * G_inverse * delta
    w = min(1, 1 / (lamda ** 0.5 * np.linalg.norm(G * w[0], 2))) * w_prime
    
    predict = 2 * (features2[selected_index].dot(w.T) >= 0) - 1
    errors_ada[t - 1] = np.mean(sentiment2[selected_index] != predict)
```

```python
import matplotlib.pyplot as plt
plt.plot(errors_ada)
plt.show()
```

```python
np.save("error_adagrad_5000.npy", errors_ada)
```
